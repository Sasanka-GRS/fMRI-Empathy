```python
def thres(A1,k,Num):
    A = A1
    A = np.real(A)
    print(A)
    N = A.shape[0]
    
    P = A[np.triu_indices(N, k=1)]
    S = np.sort(np.ravel(P))[::-1] # symmetric, upper triangle
    val = S[k]
    K = k-1
    
    while(val==0 and K>0):
        val = S[K]
        K = K-1
    
    W = A
    W[A>=val] = 1
    W[A<val] = 0

    top = A
    val1 = S[Num]
    W[A>=val1] = 10
    
    return W
```
